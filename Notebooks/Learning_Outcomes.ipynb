# **Analyzing the Impact of Engagement on Skill Development in Various Art Forms**

## **Project Overview**
This project explores the influence of attendance, participation, and leadership on skill development across various art forms, such as animation, music composition, and visual arts. Using Tableau, I created a series of visualizations to uncover key insights into how these different modes of engagement impact personal and professional growth.

The analysis focuses on:

- Overall Impact: Identifying which type of engagement has the greatest effect on skill development.
- Significance Distribution: Visualizing the frequency of statistically significant results (`*`, `**`, `***`) for each engagement type.
- Skill-Specific Impact: Highlighting which skills, such as creativity or problem-solving, are most affected by each engagement type across different art forms.

The project aims to provide actionable insights for students, educators, and institutions to better understand how different forms of engagement contribute to skill growth and inform strategies for fostering personal and professional development in creative fields.

### **Objective**
The objective of this project is to:
1. Analyze the overall impact of attendance, participation, and leadership on skill development across various art forms.
2. Visualize the distribution of statistical significance for each engagement type to determine which type of engagement (attendance, participation, or leadership) consistently shows the strongest influence on skill development.
3. Highlight the specific skills most significantly affected by engagement types through a detailed comparison of art forms and skill impact.

### **Process**
- **Data Preparation**: Consolidated and organized data across multiple art forms, focusing on the coefficient values and significance levels for attendance, participation, and leadership.
- **Visualization**: Designed and developed three key Tableau visualizations â€” a bar chart for comparing engagement impacts, a pie chart for the distribution of significance levels, and a heatmap to show skill-specific impacts across art forms.
- **Analysis**: Analyzed the visualizations to understand how different types of engagement influence skill development, identifying trends and drawing conclusions from the distribution of significance levels and the strength of engagement effects across art forms.
